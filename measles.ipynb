```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def gen_sample(beta,gamma_,len_):
    # Total population, N.
    N = 1000
    # Initial number of infected and recovered individuals, I0 and R0.
    I0, R0 = 1, 0
    # Everyone else, S0, is susceptible to infection initially.
    S0 = N - I0 - R0
    # Contact rate, beta, and mean recovery rate, gamma, (in 1/days).
    # A grid of time points (in days)
    t = np.linspace(0, len_, len_)

    # The SIR model differential equations.
    def deriv(y, t, N, beta, gamma_):
        S, I, R = y
        dSdt = -beta * S * I / N
        dIdt = beta * S * I / N - gamma_ * I
        dRdt = gamma_ * I
        return dSdt, dIdt, dRdt

    # Initial conditions vector
    y0 = S0, I0, R0
    # Integrate the SIR equations over the time grid, t.
    ret = odeint(deriv, y0, t, args=(N, beta, gamma_))
    S, I, R = ret.T
    return I
```

```python
from scipy.spatial.distance import pdist, squareform, euclidean

def distance(truth,simulated):
    s = 100
    pairwise_dists = euclidean(truth,simulated)
    K = np.exp(-pairwise_dists ** 2 / s ** 2)
    return K 


def metropolis_hastings(p,q, qdraw, nsamp, xinit,data):
    samples=np.empty((nsamp,2))
    x_prev = xinit
    accepted=0
    for i in range(nsamp):
        x_star = qdraw(x_prev)
        p_star = p(x_star,data)
        p_prev = p(x_prev,data)
        pdfratio = p_star/p_prev
        proposalratio = q(x_prev, x_star)/q(x_star, x_prev)
        if np.random.uniform() < min(1, pdfratio*proposalratio):
            samples[i] = x_star
            x_prev = x_star
            accepted +=1
        else:#we always get a sample
            samples[i]= x_prev
            
    return samples, accepted


from scipy.stats import gamma
t=10
def gammapdf(x_new, x_old):
    return gamma.pdf(x_new[0],x_old[0]*t,scale=1./t)*gamma.pdf(x_new[1],x_old[1]*t,scale=1./t)
def gammadraw(x_old):
    return [np.random.gamma(x_old[0]*t,scale=1./t),np.random.gamma(x_old[1]*t,scale=1./t)]

def density_function(beta,data):
    simulated = gen_sample(beta[0],beta[1],len(data))
    dist =  distance(simulated,data)*gamma.pdf(beta[0],1,scale=1./10)*gamma.pdf(beta[1],1,scale=1./10)
    return dist

def get_season(data):
    x0=np.random.uniform(size=2)
    samps,acc = metropolis_hastings(density_function, gammapdf, gammadraw, 1000, x0,data)
    rez = np.mean(samps,axis=0)
    return rez
```
